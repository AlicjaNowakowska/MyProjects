```python
import pandas as pd
import seaborn as sns
%store -r G
Id=[]
chance=[]
mean=[]

for i in list(G.nodes):
    Id.append(G.nodes[i]['Id'])
    chance.append(G.nodes[i]['chance'])
    mean.append(G.nodes[i]['mean'])
df=pd.DataFrame({'id':Id,'chance':chance,'mean':mean})
#df.to_csv(r'C:/Users/wojci/Documents/Studia/praca/DANE/nodes_char1000.csv',index = None, header=True)
```

```python
cm = sns.light_palette("blue", as_cmap=True)
df=df.sort_values(by='chance',ascending=False)
df.head().style.hide_index()
s = df.style.background_gradient(cmap=cm,subset=['chance','mean'])
s
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10)) 
plt.subplot(2,2,1)
plt.hist(df['mean'])
plt.title('mean out of S')
plt.ylim((0,30))
plt.subplot(2, 2, 2)
plt.hist(df['chance'])
plt.title('chance out of S')
plt.ylim((0,30))
```
